```python
import pandas as pd

gtp = [t.split('/')[1].split('.')[0] for t in a]

df_res = pd.DataFrame(b).T
df_res.columns = init.speakers
```

```python
df_res['Pred'] = df_res.idxmax(axis=1)
df_res['Truth'] = gtp
df_res['Correct'] = df_res['Pred']==df_res['Truth']
```

```python
df_res['Correct'].value_counts()
```

```python
df_res['Delta'] = np.nan
import math
for i, row in df_res.iterrows():
    sorted_val = np.sort(row[init.speakers].values)
    hi_score = sorted_val[-1]
    second_hi_score = sorted_val[-2]
    df_res.at[i, 'Delta'] = abs(hi_score-second_hi_score)
print(df_res['Delta'].describe())
m,s = df_res['Delta'].describe()[['mean','25%']]
df_res['Need-check'] = df_res['Delta'].apply(lambda x: True if x<s else False) # rows that to be checked manually
df_res['Check-correct'] = df_res['Need-check'] | df_res['Correct'] # final results after the check
print(df_res['Need-check'].value_counts())
print(df_res['Check-correct'].value_counts())
print(len(df_res.loc[df_res['Check-correct']==True])/(len(df_res)))
```
